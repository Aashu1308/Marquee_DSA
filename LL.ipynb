## Single Linked List

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
```

```python
class SingleLL:
    def __init__(self):
        self.head=None
    
    def insert_at_start(self,data):
        new_node=Node(data)
        new_node.next=self.head
        self.head = new_node
    
    def insert_at_end(self,data):
        new_node=Node(data)
        if self.head is None:
            self.head=new_node
        else:
            temp = self.head
            while temp.next:
                temp=temp.next
            temp.next = new_node
    
    def insert_at_pos(self,data,pos):
        new_node=Node(data)
        temp=self.head
        for _ in range(pos-1):
            prev = temp
            temp = temp.next
        prev.next = new_node
        new_node.next = temp
    
    def delete_at_front(self):
        if self.head is None:
            return
        self.head = self.head.next
    
    def delete_at_back(self):
        if self.head is None:
            return
        temp = self.head
        while temp.next:
            prev = temp
            temp = temp.next
        prev.next = None
    
    def delete_at_pos(self,pos):
        temp=self.head
        for i in range(pos-1):
            prev = temp
            temp = temp.next
        prev.next = temp.next

    def minmax(self):
        if self.head is None:
            return None
        else:
            m = 100000
            mx = 0
            temp = self.head
            while(temp):
                if temp.data < m:
                    m = temp.data
                if temp.data > mx:
                    mx = temp.data
                temp = temp.next
            return m,mx
        
    def search(self,val):
        temp=self.head
        if temp is None:
            return False
        while(temp):
            if temp.data == val:
                return True
            temp=temp.next
        return False
    
    def reverse(self):
        prev=None
        current=self.head
        while(current):
            nn = current.next
            current.next = prev
            prev = current
            current = nn
        self.head = prev

    
    def display(self):
        temp=self.head
        if temp:
            while(temp):
                print(temp.data,"->",end=" ")
                temp=temp.next
            print("End",end=" ")
        else:
            print("List is empty")
```

```python
s = SingleLL()
print("insert at front")
for i in range(3):
    s.insert_at_start(int(input("front insert")))
    s.display()
    print()
print("insert at end")
for i in range(3):
    s.insert_at_end(int(input("back insert")))
    s.display()
    print()
val = int(input("val"))
pos = int(input("pos"))
print(f"insert at pos (val={val} pos={pos})")
s.insert_at_pos(val,pos)
s.display()
print()
print("min,max")
m,mx=s.minmax()
print(f"min {m}, max {mx}")
val = int(input("search val"))
print(f"search val={val}")
print(s.search(val))
print("reverse")
s.reverse()
s.display()
print()
s.reverse()
print("current list:")
s.display()
print()
print("delete first element")
s.delete_at_front()
s.display()
print()
print("delete last element")
s.delete_at_back()
s.display()
print()
pos = int(input("delete pos"))
print(f"delete at pos={pos}")
s.delete_at_pos(pos)
s.display()
print()
```

### Comp Question - Phone Number Problem
1. Update phone number by name
2. Delete phone number by name
3. Reverse list

```python
class Node:
    def __init__(self,name,phone):
        self.data=name
        self.phone=phone
        self.next=None
```

```python
class PhoneLL:
    def __init__(self):
        self.head=None
    
    def insert_at_end(self,name,phone):
        new_node=Node(name,phone)
        if self.head is None:
            self.head=new_node
        else:
            temp = self.head
            while temp.next:
                temp=temp.next
            temp.next = new_node
    
    def delete_by_name(self,name):
        temp=self.head
        while temp:
            prev = temp
            temp = temp.next
            if temp.data == name:
                break
        prev.next = temp.next
        
    def update(self,name,phone):
        temp=self.head
        if temp is None:
            return
        while(temp):
            if temp.data == name:
                temp.phone = phone
                return
            temp=temp.next
    
    def reverse(self):
        prev=None
        current=self.head
        while(current):
            nn = current.next
            current.next = prev
            prev = current
            current = nn
        self.head = prev

    
    def display(self):
        temp=self.head
        if temp:
            while(temp):
                print(temp.data, temp.phone, "->",end=" ")
                temp=temp.next
            print("End",end=" ")
        else:
            print("List is empty")
```

```python
s = PhoneLL()
print("insert at end")
for i in range(4):
    s.insert_at_end(input("back insert name"),input("back insert phone"))
    s.display()
    print()

name = input("delete name")
print(f"delete at name= {name}")
s.delete_by_name(name)
s.display()
print()
print("reverse")
s.reverse()
s.display()
print()
s.reverse()
print("current list:")
s.display()
print()
val = input("update name")
phone = int(input("update phone"))
print(f"update name={val}, phone={phone}")
s.update(val,phone)
s.display()
print()
```

### Other competitive
1. binary to decimal

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
```

```python
class SingleLL:
    def __init__(self):
        self.head=None
    
    def insert_at_start(self,data):
        new_node=Node(data)
        new_node.next=self.head
        self.head = new_node

    def insert_at_end(self,data):
        new_node=Node(data)
        if self.head is None:
            self.head=new_node
        else:
            temp = self.head
            while temp.next:
                temp=temp.next
            temp.next = new_node
    
    def display(self):
        temp=self.head
        if temp:
            while(temp):
                print(temp.data,"->",end=" ")
                temp=temp.next
            print("End",end=" ")
        else:
            print("List is empty")
    
    def decimal(self,b):
        temp = b
        i=0
        s=0
        while temp:
            d=temp%10
            temp//=10
            s+=d*pow(2,i)
            i+=1
        return s
    
    def reverse(self):
        prev=None
        current=self.head
        while(current):
            nn = current.next
            current.next = prev
            prev = current
            current = nn
        self.head = prev
    
    def copy(self):
        l = SingleLL()
        temp = self.head
        while temp:
            l.insert_at_end(temp.data)
            temp = temp.next
        return l
    
    def evaluate_binary(self):
        if not self.head:
            return None
        s=""
        temp=self.head
        while temp:
            s+=str(temp.data)
            temp=temp.next
        print(f"Decimal Value via calc is {self.decimal(int(s))}")
        print(f"Decimal Value via func is {int(s,2)}")
    
    def palindrome(self):
        l=self.copy()
        l.reverse()
        temp=self.head
        temp2=l.head
        while temp:
            if temp.data!=temp2.data:
                return False
            temp=temp.next
            temp2=temp2.next
        return True
    
    def is_prime(self,n):
        if n<=1:
            return False
        if n==2:
            return True
        for i in range(2,n//2):
            if n%i==0:
                return False
        return True
    
    def prime_counter(self):
        count=0
        temp=self.head
        while temp:
            if self.is_prime(int(temp.data)):
                count+=1
            temp=temp.next
        return count

    def max_difference(self):
        if self.head is None:
            return None
        else:
            m = 100000
            mx = 0
            temp = self.head
            while(temp):
                if temp.data < m:
                    m = temp.data
                if temp.data > mx:
                    mx = temp.data
                temp = temp.next
            return mx-m
    
```

## Double Linked List

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
        self.prev=None
```

```python
class DoubleLL:
    def __init__(self):
        self.head=None
    
    def insert_at_start(self,data):
        new_node=Node(data)
        if self.head:
            new_node.next=self.head
            self.head.prev = new_node
            self.head = new_node
        else:
            self.head=new_node
    
    def insert_at_end(self,data):
        new_node=Node(data)
        if self.head is None:
            self.head=new_node
        else:
            temp = self.head
            while temp.next:
                temp=temp.next
            temp.next = new_node
            new_node.prev = temp
    
    def insert_at_pos(self,data,pos):
        new_node=Node(data)
        temp=self.head
        for _ in range(pos-1):
            prev = temp
            temp = temp.next
        prev.next = new_node
        new_node.prev = prev
        new_node.next = temp
        temp.prev = new_node
    
    def delete_at_front(self):
        if self.head is None:
            return
        self.head = self.head.next
        self.head.prev = None
    
    def delete_at_back(self):
        if self.head is None:
            return
        temp = self.head
        while temp.next:
            prev = temp
            temp = temp.next
        prev.next = None
        temp.prev = None
    
    def delete_at_pos(self,pos):
        if pos==1:
            self.delete_at_front()
            return
        temp=self.head
        for i in range(pos-1):
            prev = temp
            temp = temp.next
        prev.next = temp.next
        temp.next.prev = prev
    
    def display(self):
        temp=self.head
        if temp:
            while(temp):
                print(temp.data,"<->",end=" ")
                temp=temp.next
            print("End",end=" ")
        else:
            print("List is empty")
```

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
```

```python
class SCLL:
    def __init__(self):
        self.head=None
        self.tail=None
        self.size=0
        
    def insert_at_front(self,data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            self.head.next = self.head
            self.tail = new_node
            self.tail.next = self.head
        else:
            temp = self.head
            while temp.next != self.head:
                temp = temp.next
            temp.next = new_node
            new_node.next = self.head
            self.head = new_node
        self.size+=1
    
    def insert_at_back(self,data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            self.head.next = self.head
        else:
            temp = self.head
            while temp.next != self.head:
                temp = temp.next
            temp.next = new_node
            self.tail = new_node
            new_node.next = self.head
        self.size+=1

    def delete_at_front(self):
        if not self.head:
            return
        if self.head.next == self.head:
            self.size-=1
            return None
        temp=self.head
        nx = self.head.next
        while temp.next!=self.head:
            temp=temp.next
        temp.next = nx
        self.head = nx
        self.size-=1
    
    def delete_at_back(self):
        if not self.head:
            return
        if self.head.next == self.head:
            self.size-=1
            return None
        temp=self.head
        while temp.next!=self.head:
            prev = temp
            temp=temp.next
        prev.next = self.head
        self.tail = prev
        temp.next = None
        self.size-=1
    
    def delete_at_pos(self,pos):
        if pos==1:
            self.delete_at_front()
            return
        if pos==0:
            self.delete_at_back()
            return
        temp=self.head
        for i in range(pos-1):
            prev = temp
            temp = temp.next
        prev.next = temp.next
        self.size-=1
    
    def display(self):
        if not self.head:
            print("Empty")
        else:
            temp = self.head
            while temp.next != self.head:
                print(temp.data,"->",end=" ")
                temp=temp.next
            print(temp.data,"-> End",end="")
        # print("next of tail is",self.tail.next.data)
    
    def display_n(self,n):
        if not self.head:
            print("Empty")
        else:
            temp = self.head
            while n:
                print(temp.data,"->",end=" ")
                temp=temp.next
                if temp.data == self.head.data:
                    n-=1
            # print(temp.data,end="")
    
    def merge(self,l2):
        if not self.head:
            return l2
        temp = l2.head
        while temp.next!=l2.head:
            self.insert_at_back(temp.data)
            temp=temp.next
        self.insert_at_back(temp.data)
    
    def merge_tail(self,l2):
        if not self.head:
            return l2
        if not l2.head:
            return
        self.tail.next = l2.head
        l2.tail.next = self.head
        self.tail = l2.tail
    
    def separate(self):
        even = SCLL()
        odd = SCLL()
        if not self.head:
            return even,odd
        else:
            temp=self.head
            while temp:
                if temp.data%2==0:
                    even.insert_at_back(temp.data)
                else:
                    odd.insert_at_back(temp.data)
                temp = temp.next
                if temp.data == self.head.data:
                    break
        return even,odd
    
    def reverse(self):
        prev = None
        current = self.head
        nn = current.next
        current.next = prev
        prev = current
        current = nn
        while current != self.head:
            nn = current.next
            current.next = prev
            prev = current
            current = nn
        current.next=prev
        self.head = prev
    
    def josephus(self,n):
        k=n
        l=self.size
        if n==1:
            print("Winner is:",self.tail.data)
        else:
            while self.size>1:
                self.delete_at_pos(n)
                n = (2*n)%l
                self.display()
                print()
            print(f"Winner for {k} is: {self.head.data}")
```
